## 1) Importación de librerías y configuración

En esta sección se cargan las librerías necesarias para:
- manipulación de datos (`pandas`, `numpy`),
- visualización (`matplotlib`),
- análisis de autocorrelación (ACF/PACF),
- modelado (`AutoReg`, `SARIMAX`),
- evaluación de pronóstico (MAPE),
- y escalado de variables (`MinMaxScaler`).

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas.plotting import lag_plot, autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
```

## 2) Construcción/carga de la serie temporal de trabajo

Se crea una serie temporal de ejemplo con:
- tendencia suave,
- componente estacional,
- y ruido aleatorio.

Esto permite practicar el flujo completo de modelado aun cuando no se use un dataset externo.

```python
# Opción demo con una serie sintética + tendencia + estacionalidad
np.random.seed(42)
n = 500
t = np.arange(n)
y = 50 + 0.05*t + 5*np.sin(2*np.pi*t/24) + np.random.normal(0, 1.5, n)

df = pd.DataFrame({
    "timestamp": pd.date_range("2024-01-01", periods=n, freq="H"),
    "load": y
}).set_index("timestamp")

df.head()
```

## Análisis exploratorio
Primero se realiza un estudio descriptivo para identificar comportamiento global de la serie: tendencia, posibles ciclos y estacionalidad.


### Visualización inicial de la serie

Se grafica la serie para observar su comportamiento global (nivel, tendencia y posible estacionalidad).

```python
df["load"].plot(figsize=(14,4), title="Serie temporal (load)")
plt.show()
```

## 3) Dependencia temporal: lag plot, ACF y PACF

Estas gráficas ayudan a identificar estructura temporal:
- **Lag plot:** relación entre \(y_t\) y \(y_{t-1}\).
- **ACF:** correlación con rezagos sucesivos.
- **PACF:** correlación parcial útil para proponer orden AR.

```python
plt.figure(figsize=(5,5))
lag_plot(df["load"])
plt.title("Lag plot")
plt.show()

plt.figure(figsize=(12,4))
autocorrelation_plot(df["load"])
plt.title("Autocorrelation plot (pandas)")
plt.show()

plot_acf(df["load"], lags=48)
plt.show()

plot_pacf(df["load"], lags=48, method="ywm")
plt.show()
```

- Si hay patrón en lag plot, la serie no es aleatoria.
- ACF ayuda a observar persistencia y posibles componentes MA.
- PACF ayuda a identificar rezagos relevantes para AR (orden p).


## 4) Modelo AutoRegresivo (AR)

Se divide la serie en entrenamiento/prueba y se ajusta un modelo **AR(p)**.
Luego se generan predicciones sobre el conjunto de prueba para evaluar desempeño.

```python
# Split temporal
train_size = int(len(df)*0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

# Modelo AR(p)
p = 6  # ejemplo guiado por PACF
ar_model = AutoReg(train["load"], lags=p, old_names=False).fit()
print(ar_model.summary())
print("AIC:", ar_model.aic, "BIC:", ar_model.bic)
```

## 5) Diferenciación y estacionariedad

Se aplica primera diferencia (\(d=1\)) para reducir tendencia/no estacionariedad.
Después se vuelve a inspeccionar ACF/PACF sobre la serie diferenciada.

```python
# Primera diferencia
df["load_diff1"] = df["load"].diff()
df["load_diff1"].dropna().plot(figsize=(14,4), title="Serie diferenciada (d=1)")
plt.show()

plot_acf(df["load_diff1"].dropna(), lags=48)
plt.show()

plot_pacf(df["load_diff1"].dropna(), lags=48, method="ywm")
plt.show()
```

Si la serie original no es estacionaria, se aplica diferenciación.
En ARIMA(p,d,q):
- p: orden AR
- d: número de diferenciaciones
- q: orden MA


## 6) Modelo SARIMAX

Se entrena un modelo SARIMAX con:
- orden no estacional \((p,d,q)\),
- orden estacional \((P,D,Q,s)\),
- y configuración para permitir ajuste flexible.

También se aplica escalado como parte del flujo del ejemplo.

```python
# Escalado opcional (como en ejemplos del capítulo)
scaler = MinMaxScaler()
train_scaled = train.copy()
test_scaled = test.copy()
train_scaled["load"] = scaler.fit_transform(train[["load"]])
test_scaled["load"] = scaler.transform(test[["load"]])

# Modelo SARIMAX simple
order = (2,1,0)
seasonal_order = (1,1,0,24)  # estacionalidad horaria diaria

sarimax_model = SARIMAX(train_scaled["load"],
                        order=order,
                        seasonal_order=seasonal_order,
                        enforce_stationarity=False,
                        enforce_invertibility=False).fit(disp=False)

print(sarimax_model.summary())
```

### Pronóstico con SARIMAX

Se generan pronósticos para el horizonte de prueba, se invierte el escalado y se calcula el error MAPE.

```python
# Forecast SARIMAX
steps = len(test_scaled)
pred_scaled = sarimax_model.forecast(steps=steps)

# Inversa de escala
pred_sarimax = scaler.inverse_transform(np.array(pred_scaled).reshape(-1,1)).ravel()
mape_sarimax = mean_absolute_percentage_error(test["load"].values, pred_sarimax)
print("MAPE SARIMAX:", round(mape_sarimax*100, 2), "%")

plt.figure(figsize=(14,4))
plt.plot(test.index, test["load"], label="Real")
plt.plot(test.index, pred_sarimax, label="Pred SARIMAX")
plt.legend()
plt.title("SARIMAX Forecast")
plt.show()
```

## 7) Bloque de verificación de predicciones (AR)

Este bloque asegura que exista `pred_ar` para evitar errores por ejecución fuera de orden en Jupyter.

```python
# 1) Predicción AR (si no existe)
pred_ar = ar_model.predict(start=test.index[0], end=test.index[-1])
```

## 8) Bloque de verificación de predicciones (SARIMAX)

Este bloque asegura que exista `pred_sarimax` y la convierte a escala original cuando corresponde.

```python
# 2) Predicción SARIMAX (si no existe)
# Si ya tienes pred_sarimax en escala original, puedes saltar esta celda.
steps = len(test_scaled)
pred_scaled = sarimax_model.forecast(steps=steps)
pred_sarimax = scaler.inverse_transform(np.array(pred_scaled).reshape(-1, 1)).ravel()
```

## 9) Comparación de modelos con MAPE

Se calculan los MAPE de AR y SARIMAX y se resume el resultado en una tabla comparativa.

```python
# 3) MAPE + tabla resumen
from sklearn.metrics import mean_absolute_percentage_error
import pandas as pd
import numpy as np

mape_ar = mean_absolute_percentage_error(test["load"], pred_ar)
mape_sarimax = mean_absolute_percentage_error(test["load"].values, pred_sarimax)

resumen = pd.DataFrame({
    "Modelo": ["AR", "SARIMAX"],
    "MAPE (%)": [mape_ar * 100, mape_sarimax * 100]
})

resumen
```

- AR funciona como línea base interpretable.
- SARIMAX captura mejor estructuras con diferenciación y estacionalidad.
- El criterio cuantitativo usado para comparar fue **MAPE (%)**.
